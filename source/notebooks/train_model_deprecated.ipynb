```python
import pandas as pd
from pycaret.datasets import get_data
from pycaret.classification import *
```

```python
frames = []
for i in range (1,3):
    frame = pd.read_csv(f'data/final/depresja_{i}.txt', sep='\t')
    frame = frame.drop(['Annotation Name', 'Annotation Description', 'Annotation Tags', 'Mouse Position X [px]', 'Mouse Position Y [px]', 'Scroll Direction X', 'Scroll Direction Y', 'Content'], axis=1)
    frame['Class'] = 'Depressive'
    frames.append(frame)

control = pd.read_csv(f'data/final/kontrolni.txt', sep='\t')
control = control.drop(['Annotation Name', 'Annotation Description', 'Annotation Tags', 'Mouse Position X [px]', 'Mouse Position Y [px]', 'Scroll Direction X', 'Scroll Direction Y', 'Content', 'Eye Position Right X [mm]', 'Eye Position Right Y [mm]', 'Eye Position Right Z [mm]', 'Pupil Position Right X [px]', 'Pupil Position Right Y [px]'], axis=1)
control['Class'] = 'Control'
frames.append(control)
data = pd.concat(frames, ignore_index=True)
data.drop_duplicates(keep="first", inplace=True, ignore_index=True)
```

```python
s = setup(data, target = 'Class', session_id = 123)
```

```python
best = compare_models()
```

```python
save_model(best, 'my_best_pipeline')
```
